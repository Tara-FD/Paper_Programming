```python
import pandas as pd
import scipy.stats as stats
```

```python
df = pd.read_csv('../csv files/df_sleep_adhd.csv')
df.head()
```

## creating dichotomus for wave 4(p4) conners scale:

```python
p4_cpadhd = df[['id_number','p4_cpadhdi_raw']].dropna()
```

```python
p4_cpadhd['zscore_p4_cpadhd'] = stats.zscore(list(p4_cpadhd['p4_cpadhdi_raw'].values))
```

```python
p4_cpadhd['tscore_p4_cpadhd'] = ((p4_cpadhd['zscore_p4_cpadhd'].values)*10) + 50
```

```python
p4_cpadhd['dich_tscore_p4_cpadhd'] = ((p4_cpadhd['zscore_p4_cpadhd'].values)*10) + 50
```

```python
p4_cpadhd = p4_cpadhd.reset_index()
```

```python
for i,ind in enumerate(p4_cpadhd['dich_tscore_p4_cpadhd']):
    if ind >= 65:
        p4_cpadhd['dich_tscore_p4_cpadhd'][i] = 1
    else:
        p4_cpadhd['dich_tscore_p4_cpadhd'][i] = 0
```

```python
p4_cpadhd.columns
```

```python
condich_sleep_w4 = pd.merge(left=p4_cpadhd, right=df[['id_number', 'c4_sws_gb', 'c4_sws_ms', 'c4_sws_rs', 'c4_sws_rw', 'c4_sws_fa', 'c4_sws_total']], how='left', on='id_number')
```

```python
condich_sleep_w4.to_csv('../csv files/condich_sleep_w4.csv')
```
