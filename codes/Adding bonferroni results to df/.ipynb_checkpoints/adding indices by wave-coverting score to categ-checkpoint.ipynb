## converting sws scores to categorical

```python
df['sws_gb'] = np.nan
df['sws_fa'] = np.nan
df['sws_ms'] = np.nan
df['sws_rs'] = np.nan
df['sws_rw'] = np.nan
df['sws_total'] = np.nan
```

```python
for i,wave in enumerate(df['imaging waves']):
    if wave == 'ses-wave2':
        df.iloc[i, df.columns.get_loc('sws_gb')] = df.iloc[i, df.columns.get_loc('c4_sws_gb')]
        df.iloc[i, df.columns.get_loc('sws_fa')] = df.iloc[i, df.columns.get_loc('c4_sws_fa')]
        df.iloc[i, df.columns.get_loc('sws_ms')] = df.iloc[i, df.columns.get_loc('c4_sws_ms')]
        df.iloc[i, df.columns.get_loc('sws_rs')] = df.iloc[i, df.columns.get_loc('c4_sws_rs')]
        df.iloc[i, df.columns.get_loc('sws_rw')] = df.iloc[i, df.columns.get_loc('c4_sws_rw')]
        df.iloc[i, df.columns.get_loc('sws_total')] = df.iloc[i, df.columns.get_loc('c4_sws_total')]
        
    elif wave == 'ses-wave3':
        df.iloc[i, df.columns.get_loc('sws_gb')] = df.iloc[i, df.columns.get_loc('c5_sws_gb')]
        df.iloc[i, df.columns.get_loc('sws_fa')] = df.iloc[i, df.columns.get_loc('c5_sws_fa')]
        df.iloc[i, df.columns.get_loc('sws_ms')] = df.iloc[i, df.columns.get_loc('c5_sws_ms')]
        df.iloc[i, df.columns.get_loc('sws_rs')] = df.iloc[i, df.columns.get_loc('c5_sws_rs')]
        df.iloc[i, df.columns.get_loc('sws_rw')] = df.iloc[i, df.columns.get_loc('c5_sws_rw')]
        df.iloc[i, df.columns.get_loc('sws_total')] = df.iloc[i, df.columns.get_loc('c5_sws_total')]
```

## converting adhd scores to categorical

```python
df['cpadhd_raw'] = np.nan
```

```python
for i,wave in enumerate(df['imaging waves']):
    if wave == 'ses-wave2':
        df.iloc[i, df.columns.get_loc('cpadhd_raw')] = df.iloc[i, df.columns.get_loc('p4_cpadhdi_raw')]
        
    elif wave == 'ses-wave3':
        df.iloc[i, df.columns.get_loc('cpadhd_raw')] = df.iloc[i, df.columns.get_loc('p5_cpadhdi_raw')]
```
