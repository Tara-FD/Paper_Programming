## adding all subscales of sws in form of 4060 to dataframe

```python
df = pd.read_csv('BrainBehaviour_2June2022.csv')
df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1, inplace = True)
df.head()
```

```python
sws = ['fa','ms','rs','rw','total']
```

```python
for j,s in enumerate(sws):
    name = 'dich_tscore_sws_' + sws[j] 
    name2 = 'dich_tscore_sws_' + sws[j] 
    df[name] = np.nan
    df[name2] = np.nan
    #print(name)
```

```python
sws = ['fa','ms','rs','rw','total']
for i,wave in enumerate(df['imaging waves']):
    for j in range(len(sws)):
        if wave == 'ses-wave2':
            name1 = 'dich_sws_' + sws[j] + '_4060'
            df[name1] = np.nan
            name2 = 'tscore_c4_sws_' + sws[j]
            for k,ind in enumerate(df[name2]):
                if df[name2][k] >= 60:
                    df[name1][k] = '1'
                elif df[name2][k] <= 40:
                    df[name1][k] = '0'
    
        elif wave == 'ses-wave3':
            name1 = 'dich_sws_' + sws[j] + '_4060'
            df[name1] = np.nan
            name2 = 'tscore_c5_sws_' + sws[j]
            for k,ind in enumerate(df[name2]):
                if df[name2][k] >= 60:
                    df[name1][k] = '1'
                elif df[name2][k] <= 40:
                    df[name1][k] = '0'
```
