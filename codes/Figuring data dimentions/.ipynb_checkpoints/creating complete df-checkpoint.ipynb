### creating categorical measurement from conners with 60 and 70 cutoffs

```python
df.head()
```

```python
df = df.rename({'dich_tscore_p4_cpadhd' : 'dich_tscore_p4_cpadhd_65', 'dich_tscore_p5_cpadhd': 'dich_tscore_p5_cpadhd_65'}, axis = 1)
```

```python
df.head()
```

```python
# wave 4: 40, 60 and 70

df['dich_tscore_p4_cpadhd_40'] = df['dich_tscore_p4_cpadhd_65']
for i,ind in enumerate(df['tscore_p4_cpadhd']):
    if ind >= 40:
        df['dich_tscore_p4_cpadhd_40'][i] = 1
    else:
        df['dich_tscore_p4_cpadhd_40'][i] = 0
        
        
df['dich_tscore_p4_cpadhd_60'] = df['dich_tscore_p4_cpadhd_65']
for i,ind in enumerate(df['tscore_p4_cpadhd']):
    if ind >= 60:
        df['dich_tscore_p4_cpadhd_60'][i] = 1
    else:
        df['dich_tscore_p4_cpadhd_60'][i] = 0
        

df['dich_tscore_p4_cpadhd_70'] = df['dich_tscore_p4_cpadhd_65']
for i,ind in enumerate(df['tscore_p4_cpadhd']):
    if ind >= 70:
        df['dich_tscore_p4_cpadhd_70'][i] = 1
    else:
        df['dich_tscore_p4_cpadhd_70'][i] = 0
        
df['dich_tscore_p4_cpadhd_4060'] = df['dich_tscore_p4_cpadhd_65']
for i,ind in enumerate(df['tscore_p4_cpadhd']):
    if ind >= 60:
        df['dich_tscore_p4_cpadhd_70'][i] = 1
    elif ind <= 40:
        df['dich_tscore_p4_cpadhd_70'][i] = 0
        
#=============== wave 5: 40, 60 and 70 ===================

df['dich_tscore_p5_cpadhd_40'] = df['dich_tscore_p5_cpadhd_65']
for i,ind in enumerate(df['tscore_p5_cpadhd']):
    if ind >= 60:
        df['dich_tscore_p5_cpadhd_40'][i] = 1
    else:
        df['dich_tscore_p5_cpadhd_40'][i] = 0
        

df['dich_tscore_p5_cpadhd_60'] = df['dich_tscore_p5_cpadhd_65']
for i,ind in enumerate(df['tscore_p5_cpadhd']):
    if ind >= 60:
        df['dich_tscore_p5_cpadhd_60'][i] = 1
    else:
        df['dich_tscore_p5_cpadhd_60'][i] = 0
        

df['dich_tscore_p5_cpadhd_70'] = df['dich_tscore_p5_cpadhd_65']
for i,ind in enumerate(df['tscore_p5_cpadhd']):
    if ind >= 70:
        df['dich_tscore_p5_cpadhd_70'][i] = 1
    else:
        df['dich_tscore_p5_cpadhd_70'][i] = 0
        
df['dich_tscore_p5_cpadhd_4060'] = df['dich_tscore_p5_cpadhd_65']
for i,ind in enumerate(df['tscore_p5_cpadhd']):
    if ind >= 60:
        df['dich_tscore_p5_cpadhd_4060'][i] = 1
    elif ind <= 40:
        df['dich_tscore_p5_cpadhd_4060'][i] = 0
```

### creating categorical measurement from SDQ with 60, 65, and 70 cutoffs (t-score)

```python
df.shape
```

```python
lst = ['p4_sdqemotion', 'p4_sdqconduct','p4_sdqhyper','p4_sdqpeer','p4_sdqprosoc','p4_sdqtot','p4_sdqimpact','p5_sdqemotion','p5_sdqconduct',
       'p5_sdqhyper','p5_sdqpeer','p5_sdqprosoc','p5_sdqtot','p5_sdqimpact']
#drop = df[lst]
```

```python
df = df.dropna(axis = 0,subset = ['p4_sdqemotion', 'p4_sdqconduct','p4_sdqhyper','p4_sdqpeer','p4_sdqprosoc','p4_sdqtot','p4_sdqimpact','p5_sdqemotion','p5_sdqconduct',
       'p5_sdqhyper','p5_sdqpeer','p5_sdqprosoc','p5_sdqtot','p5_sdqimpact'])
#df = df[pd.notnull(df[['p4_sdqemotion', 'p4_sdqconduct','p4_sdqhyper','p4_sdqpeer','p4_sdqprosoc','p4_sdqtot','p4_sdqimpact','p5_sdqemotion','p5_sdqconduct',
#       'p5_sdqhyper','p5_sdqpeer','p5_sdqprosoc','p5_sdqtot','p5_sdqimpact']])]
```

```python
df.shape
```

```python
for col in df.columns:
    if col in lst:
        zscore = 'zscore_' + col
        df[zscore] = stats.zscore(list(df[col].values))
        tscore = 'tscore_' + col
        df[tscore] = ((df[zscore])*10) + 50
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.reset_index()
```

```python
df.head()
```

```python
for col in df.columns:
    if col.find("tscore_p") == 0:
        name = "dich_" + col + "_40"
        df[name] = df[col]
        for i,ind in enumerate(df[col]):
            if ind >= 40:
                df[name][i] = '1'
            else:
                df[name][i] = '0'
                
for col in df.columns:
    if col.find("tscore_p") == 0:
        name = "dich_" + col + "_60"
        df[name] = df[col]
        for i,ind in enumerate(df[col]):
            if ind >= 60:
                df[name][i] = '1'
            else:
                df[name][i] = '0'
                
                
for col in df.columns:
    if col.find("tscore_p") == 0:
        name = "dich_" + col + "_65"
        df[name] = df[col]
        for i,ind in enumerate(df[col]):
            if ind >= 65:
                df[name][i] = '1'
            else:
                df[name][i] = '0'
        
        
for col in df.columns:
    if col.find("tscore_p") == 0:
        name = "dich_" + col + "_70"
        df[name] = df[col]
        for i,ind in enumerate(df[col]):
            if ind >= 70:
                df[name][i] = '1'
            else:
                df[name][i] = '0'
                
for col in df.columns:
    if col.find("tscore_p") == 0:
        name = "dich_" + col + "_4060"
        df[name] = df[col]
        for i,ind in enumerate(df[col]):
            if ind >= 60:
                df[name][i] = '1'
            elif ind <=40:
                df[name][i] = '0'
            else:
                df[name][i] = 'NaN'
```

## Converting ASWS to cathegorical

```python
lst = ['c4_sws_gb', 'c4_sws_fa', 'c4_sws_ms', 'c4_sws_rs', 'c4_sws_rw', 'c4_sws_total',
        'c5_sws_gb', 'c5_sws_fa', 'c5_sws_ms', 'c5_sws_rs', 'c5_sws_rw', 'c5_sws_total']
df = df.dropna(axis = 0,subset = lst)
```

```python
df.shape
```

```python
for col in df.columns:
    if (col.find("c4_sws") == 0 or col.find("c5_sws") == 0):
        #print(col)
        zscore = 'zscore_' + col
        df[zscore] = stats.zscore(list(df[col].values))
        tscore = 'tscore_' + col
        df[tscore] = ((df[zscore])*10) + 50
```

```python
df.shape
```

```python
df = df.reset_index()
```

```python
for col in df.columns:
    if col.find("tscore_c") == 0:
        name = "dich_" + col + "_40"
        df[name] = df[col]
        for i,ind in enumerate(df[col]):
            if ind >= 40:
                df[name][i] = '1'
            else:
                df[name][i] = '0'
                
for col in df.columns:
    if col.find("tscore_c") == 0:
        name = "dich_" + col + "_60"
        df[name] = df[col]
        for i,ind in enumerate(df[col]):
            if ind >= 60:
                df[name][i] = '1'
            else:
                df[name][i] = '0'
                
                
for col in df.columns:
    if col.find("tscore_c") == 0:
        name = "dich_" + col + "_65"
        df[name] = df[col]
        for i,ind in enumerate(df[col]):
            if ind >= 65:
                df[name][i] = '1'
            else:
                df[name][i] = '0'
        
        
for col in df.columns:
    if col.find("tscore_c") == 0:
        name = "dich_" + col + "_70"
        df[name] = df[col]
        for i,ind in enumerate(df[col]):
            if ind >= 70:
                df[name][i] = '1'
            else:
                df[name][i] = '0'
```
